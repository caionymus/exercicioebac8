#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dados_sinasc = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
dados = dados_sinasc[variaveis]

# configurando o layout do painel de gráficos
fig, axes = plt.subplots(2, 3, figsize=(18, 8))
fig.suptitle('Gráficos de Barras para Variáveis', fontsize=16)

# loop para criar os gráficos
for i, variavel in enumerate(variaveis):
    row = i // 3
    col = i % 3
    sns.countplot(data=dados, x=variavel, ax=axes[row, col])
    axes[row, col].set_title(f'{variavel}', fontsize=14)
    axes[row, col].set_xlabel(variavel, fontsize=12)

# ajustando o layout para evitar sobreposições
plt.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media_apgar5_por_consultas = dados_sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(media_apgar5_por_consultas['CONSULTAS'], media_apgar5_por_consultas['APGAR5'])
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')
plt.title('Média de APGAR5 por CONSULTAS')

plt.xticks(rotation=45)

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
variaveis = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

# configurando o layout do painel de gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('SEXO, GESTACAO, ESTCIVMAE e CONSULTAS')

# loop para criar os gráficos
for i, variavel in enumerate(variaveis):
    row = i // 2
    col = i % 2
    sns.countplot(data=dados_sinasc, x=variavel, ax=axes[row, col])
    axes[row, col].set_title(variavel)
    axes[row, col].set_xlabel(variavel)


# ajustando o layout para evitar sobreposições
plt.tight_layout()
plt.subplots_adjust(top=0.9)

plt.show()
```
